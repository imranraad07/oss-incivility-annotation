```python
import requests
import pandas as pd

token = ''
second_token = ''
third_token = ''
headers = {"Authorization": "Bearer "}

heated = pd.read_csv('./heated_issues.csv')
```

```python
heated_list = heated.to_dict(orient='records')
heated_list[0]
```

```python
stop_flag = False
issue_dict = {
    'issue_id':[],
    'issue_number': [],
    'project_full_name':[],
    'created_at':[],
    'html_url':[],
    'api_url':[]
}
for i in range(0, len(heated_list)):
    print(i)
    repo = heated_list[i]
    page = 1
    number_of_issues = 0
    number_of_locked_issues = 0
    number_of_heated_issues = 0

    while(True):
        print('page number: '+ str(page))
        issues_url = repo.get('issues_url').replace('{/number}', '') + '?state=all&per_page=100&page=' + str(page)
        response = requests.get(issues_url, headers=headers)
        expected_number_of_heated_issues = repo.get('number_of_heated_issues')

        if response.status_code != 200:
            print(response.content)
            if "API rate limit exceeded for user" in str(response.content):
                stop_flag = True
                break
        response = response.json()
        if len(response) == 0:
            break
        number_of_issues += len(response)
        for issue in response:
            locked = issue.get('locked')
            lock_reason = issue.get('active_lock_reason')
            state_reason = issue.get('state_reason')

            if (lock_reason is not None and 'heated' in lock_reason.lower()) or (state_reason is not None and 'heated' in state_reason.lower()):
                created_at = issue.get('created_at')
                issue_id = issue.get('id')
                issue_number = issue.get('number')
                project_full_name = issue.get('full_name')
                html_url = issue.get('html_url')
                api_url = issue.get('url')
                issue_dict['issue_id'].append(issue_id)
                issue_dict['project_full_name'].append(project_full_name)
                issue_dict['created_at'].append(created_at)
                issue_dict['html_url'].append(html_url)
                issue_dict['api_url'].append(api_url)
                number_of_heated_issues += 1

        if (expected_number_of_heated_issues == number_of_heated_issues):
            break
        page += 1
    if stop_flag:
        break
```

### Extract comments from issues

```python
df = pd.read_csv('./merged_threads.csv')
df["api_url"] = df["api_url"].apply(lambda x: x.replace("/pull/", "/issues/"))
df.reset_index(drop=True, inplace=True)
df.to_csv('./merged_threads_v1.csv')
```

```python
comments_details = {
    'project_name': [],
    'issue_id': [],
    'comment_id': [],
    'comment_body': [],
    'created_at':[],
    'user_id':[]
}
bodies = []
is_ids = []
co_ids = []
names = []
users = []
created_ats = []

api_urls = df['api_url'].tolist()
issue_ids = df['issue_id'].tolist()
project_names = df['project_full_name'].tolist()
n = 0
for api in api_urls:
    page = 1
    response = requests.get(api, headers=headers)
    response = response.json()

    issue_id = issue_ids[n]
    project_name = project_names[n]

    bodies.append(response.get('body'))
    is_ids.append(issue_id)
    co_ids.append(response.get('id'))
    names.append(project_name)
    users.append(response.get('user').get('id'))
    created_ats.append(response.get('created_at'))

    while True:
        comments_url = response.get('comments_url') + '?per_page=100&page={}'.format(page)
        response_comment = requests.get(comments_url, headers=headers)
        response_comment = response_comment.json()
        for comment in response_comment:
            bodies.append(comment.get('body'))
            is_ids.append(issue_id)
            co_ids.append(comment.get('id'))
            names.append(project_name)
            users.append(comment.get('user').get('id'))
            created_ats.append(comment.get('created_at'))
        if len(response_comment) == 0:
            break
        page +=1
    n += 1
```

```python
comments_details['project_name'] = names
comments_details['issue_id'] = is_ids
comments_details['comment_id'] = co_ids
comments_details['comment_body'] = bodies
comments_details['created_at'] = created_ats
comments_details['user_id'] = users
```

```python
df_comments = pd.DataFrame(comments_details)
df_comments.head()
```
